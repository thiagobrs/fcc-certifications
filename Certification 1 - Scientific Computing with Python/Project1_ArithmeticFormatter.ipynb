# **Arithmetic Formatter**

Students in primary school often arrange arithmetic problems vertically to make them easier to solve. For example, "235 + 52" becomes:

"235"

"+ 52"

"-------"

Create a function that receives a list of strings that are arithmetic problems and returns the problems arranged vertically and side-by-side. The function should optionally take a second argument. When the second argument is set to True, the answers should be displayed.

**Example**

Function Call: `arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"])`

Output:

"   32      3801      45      123"

"+ 698    -    2    + 43    +  49"

"-----    ------    ----    -----"

# **Rules**
The function will return the correct conversion if the supplied problems are properly formatted, otherwise, it will return a string that describes an error that is meaningful to the user.

Situations that will return an error:
* If there are too many problems supplied to the function. The limit is five, anything more will return: Error: Too many problems.
* The appropriate operators the function will accept are addition and subtraction. Multiplication and division will return an error. Other operators not mentioned in this bullet point will not need to be tested. The error returned will be: Error: Operator must be '+' or '-'.
* Each number (operand) should only contain digits. Otherwise, the function will return: Error: Numbers must only contain digits.
* Each operand (aka number on each side of the operator) has a max of four digits in width. Otherwise, the error string returned will be: Error: Numbers cannot be more than four digits.

If the user supplied the correct format of problems, the conversion you return will follow these rules:
*There should be a single space between the operator and the longest of the two operands, the operator will be on the same line as the second operand, both operands will be in the same order as provided (the first will be the top one and the second will be the bottom.
* Numbers should be right-aligned.
* There should be four spaces between each problem.
* There should be dashes at the bottom of each problem. The dashes should run along the entire length of each problem individually. (The example above shows what this should look like.)

```python
# Validates the problems format and stores each term into a proper list
def validate_problems(problems, showAnswer):
  errorMsg = ''
  operands1 = [] 
  operators = [] 
  operands2 = [] 
  answers = []

  # Checks if the number of problems exceeds the maximum value allowed
  if len(problems) > 5:
    errorMsg = "Error: Too many problems."
  elif len(problems) == 0:
    errorMsg = "Error: No problems to arrange."
  else:
    # Iterate over the list of problems
    for p in problems:
      # Checks if the number of terms in the given problem is equal to 3 (operand, operator, operand)
      if len(p.split()) != 3:
        errorMsg = "Error: Too many terms in problem '" + p + "'"
        break
      else: 
        # Splits the terms of the given problem
        terms = p.split()

        # Checks if the operator is valid
        if terms[1] not in ['+', '-']:
          errorMsg = "Error: Operator must be '+' or '-'."
          break
        else:
          operator = terms[1] # Because the operator should always be the second term of a problem in the format: operand operator operand

        # Checks if the operands are numbers (integers for the porpose of this example)
        try:
          operand1 = int(terms[0])
          operand2 = int(terms[2])
        except:
          errorMsg = "Error: Numbers must only contain digits."
          break
        
        # Checks if any operand has more than 4 digits
        if (operand1 > 9999) or (operand2 > 9999):
          errorMsg = "Error: Numbers cannot be more than four digits."
          break
        else:
          # Appends the correspondents operands, operator and answer for the given problem as strings
          operands1.append(terms[0])
          operators.append(terms[1])
          operands2.append(terms[2])

          if showAnswer:
            if operator == '+':
              answers.append(str(operand1 + operand2))
            elif operator == '-':
              answers.append(str(operand1 - operand2))

  return operands1, operators, operands2, answers, errorMsg

# Converts the given problems to the proper format (arranged vertically and side-by-side)
def arithmetic_arranger(problems, showAnswer = False):
  arranged_problems = ""

  # Validates the problems format and extracts each term to a proper list
  operands1, operators, operands2, answers, errorMsg = validate_problems(problems, showAnswer)

  # Checks if there were any errors in the validation
  if errorMsg != '': 
    arranged_problems = errorMsg
  else:
    sep, line1, line2, line3, line4, size = "", "", "", "", "", 0

    # For each problem, grabs
    for i in range(len(problems)):
      # If there is more than 1 problem, separates them
      if i > 0:
        sep = "    "

      # Gets the size of the bigger operand
      if len(operands1[i]) >= len(operands2[i]):
        size = len(operands1[i])
      else: 
        size = len(operands2[i])

      # Appends the first operand in the correct position of the first line
      line1 += sep + operands1[i].rjust(size + 2)

      # Appends the operator and the second operand in the correct position of the second line
      line2 += sep + operators[i] + " " + operands2[i].rjust(size)

      # Appends the dashes in the third line to separate the problem from the answer
      line3 += sep + "".rjust(size + 2, "-")

      # If answer was requested, then appends it in the fourth line
      if showAnswer:
        line4 += sep + answers[i].rjust(size + 2)
      
    # Puts everything together in a single string
    if showAnswer:
      arranged_problems = line1 + '\n' + line2 + '\n' + line3 + '\n' + line4
    else:
      arranged_problems = line1 + '\n' + line2 + '\n' + line3

  return arranged_problems

# Examples
#print(arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49", "123 + 49", "123 + 49"]))
#print(arithmetic_arranger([]))
#print(arithmetic_arranger(["32 + 698 - 39", "3801 - 2", "45 + 43", "123 + 49"]))
#print(arithmetic_arranger(["32 x 698", "3801 - 2", "45 + 43", "123 + 49"]))
#print(arithmetic_arranger(["32a + 698", "3801 - 2", "45 + 43", "123 + 49"]))
#print(arithmetic_arranger(["32000 + 698", "3801 - 2", "45 + 43", "123 + 49"]))
#print(arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"], True))
print(arithmetic_arranger(["32 + 8", "1 - 3801", "9999 + 9999", "523 - 49"], True))
```
